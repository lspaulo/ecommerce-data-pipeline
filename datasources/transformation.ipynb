```python
import pandas as pd
```

```python
# Lendo o dataframe
fact_reviews = pd.read_csv('../woman_clothing/Womens Clothing E-Commerce Reviews.csv', index_col=0)
fact_reviews.sample(5)
```

```python
# Verificando informações do df
fact_reviews.info()
```

```python
# Lista colunas do df
list(fact_reviews.columns)
```

```python
# Remover registros com valores nulos
fact_reviews = fact_reviews.dropna(
    # O axis poder ser 0 ou 1
    # 0 deleta a linha e 1 a coluna
    axis=0,
    subset=['Clothing ID',
            'Title',
            'Review Text',
            'Rating']
)

fact_reviews
```

```python
fact_reviews.info()
```

```python
# Normalização de texto
# Remove caracteres que não são alfanuméricos (letras e números) ou espaços.
def std_text(text):
    text = str(text)
    cleaned_text = ''.join(e for e in text if e.isalnum() or e.isspace())
    cleaned_text = cleaned_text.lower()

    return cleaned_text
```

```python
# chamando a funão criada:
fact_reviews["Title"] = fact_reviews["Title"].apply(lambda x: std_text(x))
fact_reviews["Review Text"] = fact_reviews["Review Text"].apply(lambda x: std_text(x))
```

```python
# Visualizando novo df
fact_reviews
```

```python
# Carregando um segundo dataset com as categorias das roupas
dim_categories = pd.read_csv('../woman_clothing/womens_departments.csv', index_col=0)
dim_categories
```

```python
dim_categories.info()
```

```python
# Removendo valores nulos
dim_categories = dim_categories.dropna(axis=0)
```

```python
dim_categories.info()
```

```python
# Fazendo o join entre as duas tabelas
merged_df = pd.merge(
    fact_reviews,
    dim_categories,
    # A Chave que vamos utilizar, um valor que temos nas duas tabelas
    on="Clothing ID",
    # Tipo de merge que vamos fazer no caso aqui "inner"
    how="inner"
)
```

```python
merged_df.sample(10)
```

```python
merged_df.info()
```

```python
# Verificar valores da coluna Rating
merged_df["Rating"].unique()
```

```python
# Pegando somente os valores que são núemros
merged_df = merged_df[merged_df["Rating"].isin(['3', '5', '2', '4', '1'])]
```

```python
merged_df["Rating"].unique()
```

```python
# Modificar o tipo do dado
merged_df['Rating'] = merged_df['Rating'].astype(int)
```

```python
merged_df.info()
```

```python
# Agrupar por categoria e calcular a media de rating
avg_rating = merged_df.groupby("Class Name")["Rating"].mean()

avg_rating
```

```python
df_final_rating = pd.DataFrame(avg_rating).sort_values("Rating", ascending=False)
```

```python
df_final_rating.round(1)
```
